## Install the relevant libraries

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
standard_url='https://fbref.com/en/comps/9/Premier-League-Stats'
```

```python
data=requests.get(standard_url)
```

```python
soup=BeautifulSoup(data.text)
```

## Now let's do shooting tables ,standing tables and other general tables

### Standings Table

```python
standings=pd.read_html(data.text,match='Regular season Table')[0]
```

```python
standings.drop('Notes',axis=1,inplace=True)
```

```python
standings
```

## Shooting tables




```python
shooting=pd.read_html(data.text,match='Squad Shooting')[0]
```

```python
shooting
```

### Passing Tables


```python
passing=pd.read_html(data.text,match='Squad Passing')[0]
passing.head()
```

```python
pass_types=pd.read_html(data.text,match='Squad Pass Types')[0]
pass_types.head()
```

### Defensive Stats Tables

```python
defence=pd.read_html(data.text,match='Squad Defensive Actions')[0]
defence.head()
```

## Posession Stats

```python
possession=pd.read_html(data.text,match='Squad Possession')[0]
possession.head()
```

```python
possession.columns
```

## MERGE THE DATAFRAMES

```python
pl_stats=pd.concat([possession,passing,pass_types,shooting,defence],axis=1,join='outer')
```

```python
pl_stats
```

```python
pl_stats=pd.concat([standings,pl_stats],axis=1,join='outer')
```

```python
pl_stats.drop(axis=1,columns=('Unnamed: 0_level_0',   'Squad'),inplace=True)
```

```python
pl_stats
```

```python
pl_stats.drop(19,inplace=True)
```

```python
pl_stats.set_index('Rk',inplace=True)
```
